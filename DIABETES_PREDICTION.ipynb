```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data_set=pd.read_csv('/content/diabetes.csv')
```

```python
data_set.head()
```

```python
data_set.shape
```

```python
data_set.info()
```

```python
data_set.isnull().sum()
```

```python
#gives statistical data of dataset
data_set.describe()
```

```python
data_set['Outcome'].value_counts()
```

0---> non diabetic
1---> diabetic

```python
#calculating the mean for target
data_set.groupby('Outcome').mean()
```

```python
#seperating the features and targets
features=data_set.drop(columns='Outcome',axis=1)
target=data_set['Outcome']
```

```python
print(features)
print(target)
```

DATA STANDARDIZATION

```python
scaler=StandardScaler()
scaler.fit(features)
```

```python
standardized_data=scaler.transform(features)
standardized_data
```

```python
features=standardized_data
target=data_set['Outcome']
```

```python
pd.DataFrame(features)
```

SPLITING THE DATA INTO TRAINING AND TESTING

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.2,random_state=2)
```

```python
features.shape,X_train.shape,X_test.shape
```

```python
target.shape,Y_train.shape,Y_test.shape
```

USING LOGICTIC REGRESSION MODEL

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,training_data_prediction)
print("Accuracy score of training data with logistic Regression : ",training_data_accuracy)
```

```python
testing_data_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(Y_test,testing_data_prediction)
print("Accuracy score of testing data with logistic Regression : ",testing_data_accuracy)
```

USING SVC MODEL

```python
model1=SVC(kernel='linear')
```

```python
model1.fit(X_train,Y_train)
```

```python
training_data_prediction=model1.predict(X_train)
```

```python
training_data_accuracy=accuracy_score(Y_train,training_data_prediction)
print("Accuracy score of training data with SVM : ",training_data_accuracy)
```

```python
testing_data_prediction=model1.predict(X_test)
testing_data_accuracy=accuracy_score(Y_test,testing_data_prediction)
print("Accuracy score of testing data with SVM : ",testing_data_accuracy)
```

MODEL PREDICTION WITH OUR DATA

```python
input_data=(4,110,92,0,0,37.6,0.191,30)
#changing the input data into numpy array
input_data_asnumpy_array=np.asarray(input_data)
#reshaping the array for one instance bcoz if we didnot reshape it will think we will give 768 rows if data
input_data_reshaped=input_data_asnumpy_array.reshape(1,-1)
#standardizing the data
std_data=scaler.transform(input_data_reshaped)
print(std_data)
```

```python
prediction=model1.predict(std_data)
print(prediction)
if (prediction[0]==0):# output from model is list thatswhy we need to mention index
  print("The person is not diabetic")
else:
  print("The person is diabetic")
```

PREDICTION WITH ANOTHER METHOD

```python
new_data=[[5,166,72,19,175,25.8,0.587,51]]
new_data1=scaler.transform(new_data)
prediction=model1.predict(new_data1)
print(prediction)
if(prediction[0]==0):
  print("person is non-diabetes")
else:
  print("person is diabetes")
```
